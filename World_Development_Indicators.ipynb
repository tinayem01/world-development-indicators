# **Function to Query Data**

```
def get_country_indicator_results(df,country_name,indicator_name, years = years):
  fdi = df.loc[df['Indicator Name'].str.contains(indicator_name)]
  country_indicator = fdi.loc[fdi['Country Name'] == country_name]
  country_indicator_melt = country_indicator.melt(['Indicator Name'],years)
  country_indicator_df = country_indicator_melt.pivot_table(columns = 'Indicator Name',index = 'variable',values='value',aggfunc='mean')
  return country_indicator_df
```

```
def get_summary_statistics(df,country,feature):
  mean_rate = np.mean(df[feature])
  print(f'The mean {feature} of {country}  is {np.round(mean_rate,4)}')

  max_rate = np.max(df[feature])
  max_year = df.loc[df[feature] == max_rate][feature].index[0]
  print(f'The max {feature} of {country} is {np.round(max_rate,4)}, it was obtained in {max_year}')

  min_rate = np.min(df[feature])
  min_year = df.loc[df[feature] == min_rate][feature].index[0]
  print(f'The min {feature} of {country} is {np.round(min_rate,4)}, it was obtained in {min_year}')
```
